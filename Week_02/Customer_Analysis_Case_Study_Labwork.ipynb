```python
list_gender = []

for gender in list(CA_df['gender']):
    if type(gender) == float:
        list_gender.append(gender)
    elif gender.startswith('f'):
        list_gender.append('Female')
    elif gender.startswith('F'):
        list_gender.append('Female')
    elif gender.startswith('m'):
        list_gender.append('Male')
    elif gender.startswith('M'):
        list_gender.append('Male')
    
    
```

```python
list_gender
```

```python
CA_df['gender'] = list_gender
```

```python
# divide by 100 because original series was in percentage points

divided_by_100 = []

for divide in list(CA_df['customer lifetime value']):
    divide = divide/100
    divided_by_100.append(divide)
```

```python
divided_by_100
```

```python
CA_df['customer lifetime value'] = divided_by_100
```
